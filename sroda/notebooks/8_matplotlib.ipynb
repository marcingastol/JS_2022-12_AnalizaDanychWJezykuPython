```python
#%pip install pandas
#%pip install matplotlib

import pandas as pd
import numpy as np
import matplotlib.pyplot as plot
from matplotlib.ticker import FormatStrFormatter
```

## MatPlotLib

```python
a = np.arange(100)

plot.plot(a)
```

```python
b = plot.figure()

c = b.add_subplot(2, 2, 1)
d = b.add_subplot(2, 2, 2)
e = b.add_subplot(2, 2, 3)
f = b.add_subplot(2, 2, 4)


c.scatter(np.arange(100), np.arange(100) + 10 * np.random.standard_normal(100))
d.hist(np.random.standard_normal(100).cumsum(), color="black", alpha=0.1)
f.plot(np.random.standard_normal(100).cumsum(), color="black", linestyle="dashed")
```

```python
a, b = plot.subplots(2,2, sharex=True, sharey=True)

for x in range(2):
    for y in range(2):
        b[x,y].hist(np.random.standard_normal(2000), bins=50, color="black", alpha=0.3)

a.subplots_adjust(wspace=0, hspace=0)
```

```python
b = plot.figure()
z = b.add_subplot()
z.plot(np.random.standard_normal(100).cumsum(), color="black", linestyle="dashed", marker="o");
```

```python
a = plot.figure()
b = a.add_subplot()
dane = np.random.standard_normal(10).cumsum()

b.plot(dane, color="red", linestyle="dashed", label="Default")
b.plot(dane, color="black", linestyle="dashed", drawstyle="steps-post", label="steps-post")

b.legend()
```

```python
a, b = plot.subplots()

b.plot(np.random.standard_normal(10000).cumsum())
b.set_xticks([0,2500,5000,7500,10000])
b.set_xticklabels(["sdasdfsdfsdasd","grgrsdfsdfegerg","sdssgssdfdsfg","vzxxxzvsdfcxcv","gergsdfsdfergrgg"],rotation=20,fontsize=7);

b.set_title("Moj wykres");
b.set_xlabel("Moja os x");
```

```python
a, b = plot.subplots()

b.plot(np.random.standard_normal(100).cumsum(), color="black", label="a")
b.plot(np.random.standard_normal(100).cumsum(), color="black", linestyle="dashed", label="b")
b.plot(np.random.standard_normal(100).cumsum(), color="black", linestyle="dotted", label="c")

b.xaxis.set_major_formatter(FormatStrFormatter("%.2f"))

b.legend()
```
