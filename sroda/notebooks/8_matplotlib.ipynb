## MatPlotLib

```python
a = np.arange(100)

plot.plot(a)
```

```python
b = plot.figure()

c = b.add_subplot(2, 2, 1)
d = b.add_subplot(2, 2, 2)
e = b.add_subplot(2, 2, 3)
f = b.add_subplot(2, 2, 4)


c.scatter(np.arange(100), np.arange(100) + 10 * np.random.standard_normal(100))
d.hist(np.random.standard_normal(100).cumsum(), color="black", alpha=0.1)
f.plot(np.random.standard_normal(100).cumsum(), color="black", linestyle="dashed")
```

```python
a, b = plot.subplots(2,2, sharex=True, sharey=True)

for x in range(2):
    for y in range(2):
        b[x,y].hist(np.random.standard_normal(2000), bins=50, color="black", alpha=0.3)

a.subplots_adjust(wspace=0, hspace=0)
```
